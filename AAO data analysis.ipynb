```python
import numpy as np
import pandas as p
import matplotlib.pyplot as plt
import os
from matplotlib.widgets import Button
```

```python
def create_ddic(path):
    """Input path as a string
       The function will walk across path and unpack any CSV files into a dictionary of arrays with file names as the key"""
    files = []
    filenames=[]
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if '.CSV' in file:
                files.append('%s/%s' %(path,os.path.join(file)))
                filenames.append('%s' %(os.path.join(file)))
    adict={}
    for count,i in enumerate(files):
        adict[count] = p.read_csv(i,skiprows=33)
    return adict
```

```python
def create_ndic(name_list):
    adict={}
    length = np.arange(0,len(name_list),1)
    for i in length:
        adict[i]=name_list[i]
    return adict
```

```python
name_list =['Waveguide 1','dead','dead','dead','dead','Waveguide 2','Waveguide 2, Light off','Waveguide 3','Waveguide 4','Waveguide 5','Waveguide 6','Waveguide 6, Light off']
```

```python
AAO1 = create_ddic('AAOdata1')
AAO2 = create_ddic('AAOdata2')
```

```python
AAO2NL = create_ndic(name_list)
```

```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import SpanSelector

fig, (ax1, ax2) = plt.subplots(2, figsize=(8, 6))

x=np.arange(1500,1600.004,0.004)
y=[]
for i in AAO2[10][AAO2[10].columns[0]]:
    y.append(i)

ax1.plot(x, y, '-')
ax1.set_title('Press right mouse button and drag to test')

line2, = ax2.plot(x, y, '-')

locs={}
class Index(object):
    ind = 0
    def prev(self,event):
        self.ind +=1
        i=self.ind
#         print('lemon! %s'%i)
        locs[i] = np.asarray(plt.ginput(2))
        plt.draw()

callback = Index()
axprev = plt.axes([0.75, 0.9, 0.125, 0.075])
bprev = Button(axprev, 'Select Points')
bprev.on_clicked(callback.prev)

def onselect(xmin, xmax):
    indmin, indmax = np.searchsorted(x, (xmin, xmax))
    indmax = min(len(x) - 1, indmax)

    thisx = np.asarray(x[indmin:indmax])
    thisy = np.asarray(y[indmin:indmax])
    line2.set_data(thisx, thisy)
    ax2.set_xlim(thisx[0], thisx[-1])
    ax2.set_ylim(thisy.min(), thisy.max())
    fig.canvas.draw()

# Set useblit=True on most backends for enhanced performance.
span = SpanSelector(ax1, onselect, 'horizontal', useblit=True,
                    rectprops=dict(alpha=0.5, facecolor='red'),button=3)



plt.show()
```

```python
"""This plotting function is now considered obselete for analysis"""
def plotting(ddic,ndic):
    length = np.arange(0,len(ddic),1)
    for i in length:
        if ndic[i] != 'dead':
            plt.plot(ddic[i])
            plt.title('%s' %(ndic[i]))
            plt.xlabel('Wavelength(nm)')
            plt.ylabel('intensity(dBm)')
            plt.show()
        elif ndic[i] == 'dead':
            print('Dead data')

plotting(AAO2,AAO2NL)
```
